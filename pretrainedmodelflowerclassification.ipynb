```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import PIL.Image as image
import tensorflow as tf
import tensorflow_hub  as hub
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import  Sequential
!pip install tf_keras
import os
os.environ["TF_USE_LEGACY_KERAS"] = "1"
```

```python
dataurl='https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz'
datadir=tf.keras.utils.get_file('flower_photos',origin=dataurl,cache_dir='.',untar=True)
datadir
import pathlib
datadir=pathlib.Path(datadir)
```

```python
list(datadir.glob('*/*/*.jpg'))
```

```python
list(datadir.glob('*/roses/*'))
```

```python
flowersdic={
    'roses':list(datadir.glob('*/roses/*')),
    'daisy':list(datadir.glob('*/daisy/*')),
    'dandelion':list(datadir.glob('*/dandelion/*')),
    'sunflowers':list(datadir.glob('*/sunflowers/*')),
    'tulips':list(datadir.glob('*/tulips/*'))
}
```

```python
flowerlabels={
    'roses':0,
    'daisy':1,
    'dandelion':2,
    'sunflowers':3,
    'tulips':4
}
```

```python
a=[]
b=[]
for x,y in flowersdic.items():
  for image in y:
    img=cv2.imread(image)
    resise=cv2.resize(img,(224,224))
    a.append(resise)
    b.append(flowerlabels[x])
```

```python
a=np.array(a)
b=np.array(b)
```

```python
a.shape
```

```python
b.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(a,b,random_state=0)
```

```python
xtrain=xtrain/255
xtest=xtest/255
```

```python
feature='https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'
pretrainedmodel=hub.KerasLayer(
    feature,
    input_shape=(224,224,3),
    trainable=False
)
import tf_keras
```

```python
import tf_keras  # important

feature_vector_url = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"

model = tf_keras.Sequential([
    hub.KerasLayer(feature_vector_url, input_shape=(224, 224, 3), trainable=False),
    tf_keras.layers.Dense(5, activation='softmax')
])

model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)
```

```python
model.fit(xtrain,ytrain,epochs=5)
```

```python
model.evaluate(xtest,ytest)
```

```python
ypredicted=model.predict(xtest)
```
